
Expectation: A working solution where the learned class label using Logistic Regression is similar (for most possible points) to the training class label. 

Optional: Show the decision boundary. Note that it is not trivial to the draw the decision boundary in the examples below. 



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plotClass(X, y, p):
    plt.figure()
    for i in range(y.shape[1]):
        if y[0, i] == 0:
            plt.plot(X[0, i], X[1, i], 'r' + p)
        else:
            plt.plot(X[0, i], X[1, i], 'b' + p)

    plt.show()


def sigmoid(z):
    return 1 / (1 + np.exp(-z))


def loss(yhat, y):
    return np.sum(-y * np.log(yhat) - (1 - y) * np.log(1 - yhat))


```

```python
def logistic_reg(X, y):
    # loss
    print(X.shape)
    print(y.shape)
    lr = 1e-3
    theta = np.random.randn(2, 1)  # initialize paramaters, D=2
    ls = []
    for i in range(1000):
        yhat = sigmoid(X @ theta)
        l = loss(yhat, y)

        ls.append(l)

        # gradient step: grad = dL/dtheta
        grad = X.T @ (yhat - y)
        theta -= lr * grad
    plt.plot(ls)
```

```python
#Q1
num_data = 1000  # data points per class

X = np.random.uniform(-1, 1, [2, num_data])
y = X[0, :] ** 2 + X[1, :] ** 2 - 0.5 > 0
y = (y[None, :]).astype(np.int8)
plotClass(X, y, 'o')
plotClass(X ** 2, y ** 2, 'x')
X_new = X ** 2
X_new = X_new.reshape(-1, 2)
Y_new = y ** 2
Y_new = Y_new.reshape(-1, 1)
```
